```python
import matplotlib.pyplot as plt
from matplotlib.axes import Axes
import numpy as np
from os import path
```

```python
def generar_graficas(archivo: str, variable: str, limites = ((45, 50), (49.995, 50.002), None, None), plot = Axes.plot, x = None, output_dir: str = None):
    y_labels = ["Aptitud promedio de la población", "Aptitud promedio del mejor individuo", "Promedio de iteraciones", "Tiempo de ejecución promedio (segundos)"]
    nombres = ["aptitud_poblacion", "aptitud_individuo", "iteraciones", "tiempo"]
    datos = np.loadtxt(archivo)
    if x is None:
        x = datos[:, 0]
    resultado = []
    for i, (y_label, limite, nombre) in enumerate(zip(y_labels, limites, nombres), 1):
        fig = plt.figure()
        ax = fig.subplots()
        plot(ax, x, datos[:, i])
        ax.set_xlabel(variable)
        ax.set_ylabel(y_label)
        ax.grid(linestyle=":")
        ax.set_ylim(limite)
        if output_dir is not None:
            out_filename = path.basename(archivo)
            out_filename = path.splitext(out_filename)[0]
            out_filename += "_" + nombre + ".png"
            out_filename = path.join(output_dir, out_filename)
            fig.savefig(out_filename)
            
        resultado.append(fig)

    return resultado
```

```python
generar_graficas("variacion_tam_pop.txt", "Tamaño de la población", output_dir="./graficas")
```

```python
generar_graficas("./datos_ejecuciones/variacion_probabilidad_cruza_1.txt", "Probabilidad de cruza", [(45, 50), (49.995, 50.002), (0, 100), None], output_dir="./figuras")
generar_graficas("./datos_ejecuciones/variacion_probabilidad_cruza_2.txt", "Probabilidad de cruza", [(45, 50), (49.995, 50.002), (0, 100), None], output_dir="./figuras")
```
